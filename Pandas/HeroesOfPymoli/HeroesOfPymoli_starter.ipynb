```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
#Get total number of players
total_players=len(purchase_data_df['SN'].unique())

#Create Total Players Data Frame
players_df = pd.DataFrame({"Total Players" : [total_players]})

players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate purchase metrics
total_items=len(purchase_data_df['Item ID'].unique())
average_price=purchase_data_df['Price'].mean()
total_purchases=len(purchase_data_df['Purchase ID'].unique())
total_revenue=purchase_data_df['Price'].sum()

#Create purchase analysis Data Frame
purch_analysis_df = pd.DataFrame({"Number of Unique Items": [total_items],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [total_purchases],
                                    "Total Revenue": [total_revenue]})

#Format Data Frame
purch_analysis_df_clean=purch_analysis_df
purch_analysis_df_clean["Average Price"] = purch_analysis_df["Average Price"].map("${:.2f}".format)
purch_analysis_df_clean["Total Revenue"] = purch_analysis_df["Total Revenue"].map("${:.2f}".format)

purch_analysis_df_clean
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Group by gender
by_gender_group = purchase_data_df.groupby("Gender")

#Get unique counts and percentages
gender_count= by_gender_group.nunique()["SN"]
gender_percent = gender_count/ total_players

#Create gender demographis Data Frame
gender_percent_df = pd.DataFrame({"Total Count": gender_count, 
                                    "Percentage of Players": gender_percent})

#Format Data Frame
gender_percent_df_clean=gender_percent_df
gender_percent_df_clean.index.name = None
gender_percent_df_clean = gender_percent_df.sort_values(by = "Total Count", ascending = False) 
gender_percent_df_clean['Percentage of Players']=gender_percent_df['Percentage of Players'].map('{:.2%}'.format)

gender_percent_df_clean
```


## Purchasing Analysis (Gender)

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Setup bins and group names
bins=[0,9,14,19,24,29,34,39,150]
group_names=["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Bin ages 
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names)

#Group by Age Group
by_age_group=purchase_data_df.groupby('Age Group')

#Get unique counts and percentages
age_count=by_age_group['SN'].nunique()
age_percent=age_count/total_players

#Create Age Demographics Data Frame
age_group_df = pd.DataFrame({"Total Count": age_count, 
                                "Percentage of Players": age_percent})

#Format Data Frame
age_group_df_clean=age_group_df
age_group_df_clean.index.name = None
age_group_df_clean['Percentage of Players']=age_group_df['Percentage of Players'].map('{:.2%}'.format)


age_group_df_clean
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group by Screen Name
by_sn_group = purchase_data_df.groupby("SN")

#Get unique counts and values
purch_count_sn=by_sn_group['Purchase ID'].count()
avg_purch_price_sn=by_sn_group['Price'].mean()
total_purch_value_sn=by_sn_group['Price'].sum()

#Create top spenders Data Frame
purch_sn_df = pd.DataFrame({"Purchase Count": purch_count_sn, 
                                "Average Purchase Price": avg_purch_price_sn,
                                "Total Purchase Value": total_purch_value_sn})

#Format Data Frame
purch_sn_df_clean=purch_sn_df
purch_sn_df_clean.index.name = None
purch_sn_df_clean = purch_sn_df.sort_values(by = "Total Purchase Value", ascending=False).head(5)
purch_sn_df_clean['Average Purchase Price']=purch_sn_df['Average Purchase Price'].map('${:.2f}'.format)
purch_sn_df_clean['Total Purchase Value']=purch_sn_df['Total Purchase Value'].map('${:.2f}'.format)

purch_sn_df_clean
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Create Data Frame with Item ID, Item Name, and Price
item_data_df=purchase_data_df[['Item ID', 'Item Name', 'Price']]

# #Group by Item ID and Item Name
item_data_group = item_data_df.groupby(['Item ID', 'Item Name'])

# #Get Unique Counts and Values
purch_count_item=item_data_group['Price'].count()
total_purch_value_item=item_data_group['Price'].sum()
item_price=total_purch_value_item/purch_count_item

#Create Popular Items Data Frame
purch_item_df = pd.DataFrame({"Purchase Count": purch_count_item,
                              "Item Price": item_price,
                              "Total Purchase Value": total_purch_value_item})

#Format Data Frame
purch_item_df_clean=purch_item_df
purch_item_df_clean = purch_item_df.sort_values(by = "Purchase Count", ascending=False).head(5)
purch_item_df_clean['Item Price']=purch_item_df['Item Price'].map('${:.2f}'.format)
purch_item_df_clean['Total Purchase Value']=purch_item_df['Total Purchase Value'].map('${:.2f}'.format)

purch_item_df_clean
```
